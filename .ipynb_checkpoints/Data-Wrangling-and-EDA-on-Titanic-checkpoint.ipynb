# Data Cleaning/data Wrangling

```python
# To generate the mean of Age
ageMean = round(df["Age"].mean(),0)
```

```python
ageMean
```

```python
# To temporarily fill the blank age rows with the age mean generated above
df["Age"].fillna(ageMean)
```

```python
# You will notice the age blank space is still showing, 263 is NAN
df["Age"].isna().sum()
```

```python
# To permanently fill the blank age rows with the age mean generated 
df["Age"].fillna(ageMean, inplace = True)
```

```python
df["Age"].isna().sum()
```

```python
# Taking care of the missing value in Fare
fare_mean = round(df['Fare'].mean(),0)
```

```python
fare_mean
```

```python
df["Fare"].fillna(fare_mean, inplace = True)
```

```python
df["Fare"].isna().sum()
```

```python
# TO drop the Cabin column from the data 
df.drop(columns = ["Cabin"], inplace = True)
```

```python
df
```

```python
embarked_mode = df["Embarked"].mode(),[0]
```

```python
embarked_mode
```

```python
df["Embarked"].fillna('embarked_mode',inplace = True)

# Note that, because the data type is in text, the embarked mode has to be in quotation 
```

```python
df["Embarked"].isna().sum()
```

```python
# TO split the MR or Mrs Title from the Name column
title_split = lambda x:x.split()[1].strip(".")
```

```python
# TO create a new column "Title" and using the splited title 
df["Title"] = df["Name"].apply(title_split)
```

```python
df
```

```python
df['Title'].value_counts()
```

```python
title_list = ["Mr", "Miss","Mrs","Master","Dr","Rev"]
```

```python
for title in df["Title"]:
    if title not in title_list:
        df["Title"].replace(title,"Mr",inplace= True)
```

```python
df["Title"].value_counts()
```

# Exploratory Data Analysis (EDA)

```python
# Unit Variante Analysis - comparing just a column
count_class = df["Pclass"].value_counts()
```

```python
count_class
```

```python
count_class.index= ["Third","First","Second"]
```

```python
count_class
```

```python
count_class.plot(kind = "bar", color= "black")
plt.title("Count of Class")
plt.xlabel("Class count")
plt.ylabel("Count")
plt.xticks(rotation = 35, color = "blue")
plt.xticks(rotation = 35, color = "green")
plt.show()
```

```python
count_Sex = df["Sex"].value_counts()
```

```python
count_Sex
```

```python
count_Sex.plot(kind = "bar", color = 'black')
plt.title("Gender count")
plt.xlabel("Count of sex")
plt.ylabel("Count")
plt.xticks(rotation = 30, color= "green")
plt.yticks(rotation = 30, color = "blue")
plt.show
```

```python
count_Sex.plot(kind = "pie")
```

```python
# To find the types or kinds of plot
help(pd.DataFrame.plot)
```

```python
# Bi Variante Analysis - comparing more than 1 column
dead = df.query("Survived ==0")["Pclass"].value_counts()
alive = df.query("Survived == 1")["Pclass"].value_counts()
```

```python
dead
```

```python
alive
```

```python
survivedStatus = pd.concat([dead,alive],axis = 1, keys=["dead","alive"])
```

```python
survivedStatus
```

```python
survivedStatus.plot(kind="bar")
plt.title("Survival Status")
plt.xlabel("Count of Survivals")
plt.ylabel("Count")
plt.xticks(rotation = 30, color= "green")
plt.yticks(rotation = 30, color = "blue")
plt.show
```
